```python
def bookls():
    # Tkinter window
    r = tk.Tk()
    r.title('BOOK LIST')
    r.geometry('600x500')
    r.config(bg='black')
    
    # Heading Label (Centered)
    l0 = tk.Label(r, text='BOOK LIST', font=('Times New Roman', 30, 'bold'), fg='white', bg='black')
    l0.grid(row=0, column=0, columnspan=2, pady=20)  # Centered heading
    
    # Scrollbar
    scroll_bar = tk.Scrollbar(r)   
    scroll_bar.grid(row=2, column=3, sticky="ns")   # Attach scrollbar to the right
    
    # Listbox with Scrollbar
    mylist = tk.Listbox(r, yscrollcommand=scroll_bar.set, width=62, height=20, bg='black', fg='white')
    mylist.grid(row=2, column=0, columnspan=2, padx=10, pady=5)
    
    # Column Headings
    isbn_heading = tk.Label(r, text="ISBN", font=('Arial', 14, 'bold'), fg='white', bg='black', width=20, anchor='w')
    title_heading = tk.Label(r, text="Book Title", font=('Arial', 14, 'bold'), fg='white', bg='black', width=20, anchor='w')
    
    isbn_heading.grid(row=1, column=0, padx=5, pady=5)
    title_heading.grid(row=1, column=1, padx=5, pady=5)
    try:
        cur.execute("SELECT ISBN, `Book-Title` FROM books")
        res = cur.fetchall()
    
        # Display book details inside Listbox
        for i in res:
              mylist.insert(tk.END, f"{str(i[0]).ljust(30)}  {i[1]}")  # Insert ISBN and Book Title
    
    except Exception as e:
        print("Error fetching data:", e)  # Handle potential errors properly
    
    
    # Back Button (Centered at Bottom)
    back_button = tk.Button(r, text='BACK', bg='white', fg='black', font=('Arial', 12, 'bold'), command=r.destroy)
    back_button.place(x=450, y=450, width=80, height=30)  # Centered horizontally at the bottom
    
    # Link scrollbar to Listbox
    scroll_bar.config(command=mylist.yview)  
    
    r.mainloop()
```

```python
def fine_checking():
    
    fine_window = tk.Tk()
    fine_window.title("Fine Records - Deleted Records")
    fine_window.geometry('500x600')
    fine_window.config(bg='black')
    
    # Fetching records from the 'returned_user' table
    cur.execute('SELECT * FROM returned_user')
    res = cur.fetchall()
    
    # Loop through fetched records and display them in labels
    k = 1
    for i in res:
        # Creating labels for displaying each record
        member_id_label = tk.Label(fine_window, text=f"Member ID: {i[0]}", font=("Arial", 12), fg="white", bg="black")
        member_id_label.grid(row=k, column=0, padx=10, pady=5, sticky='w')
    
        fine_label = tk.Label(fine_window, text=f"Fine: ₹{i[1]}", font=("Arial", 12), fg="red", bg="black")
        fine_label.grid(row=k, column=1, padx=30, pady=5, sticky='w')
    
        k += 1  # Increment row number
    
    # Function to check fines
    def fine_check():
        J = tk.Toplevel(fine_window)  # Create a new top-level window
        J.title("Fine")
        J.geometry('300x200')
    
        # Label and Entry for member_id
        member_id_label = tk.Label(J, text="Enter Member ID:", font=("Arial", 12))
        member_id_label.grid(row=0, column=0, padx=10, pady=5, sticky='w')
    
        member_id_entry = tk.Entry(J, font=("Arial", 12))
        member_id_entry.grid(row=0, column=1, padx=10, pady=5)
    
        # Label to display results
        result_label = tk.Label(J, text="", font=("Arial", 12), fg="blue")
        result_label.grid(row=1, column=0, columnspan=2, padx=10, pady=5)
    
        # Function to check fine
        def check_fine():
            member_id = member_id_entry.get()
            if not member_id.isdigit():
                result_label.config(text="Invalid Member ID.\nPlease enter a number.", fg="red")
                return
    
            cur.execute("SELECT COUNT(*) FROM returned_user WHERE member_id = %s", (member_id,))
            res = cur.fetchone()
            if res[0] == 1:
                cur.execute("SELECT fine FROM returned_user WHERE member_id = %s", (member_id,))
                res1 = cur.fetchone()
                fine_amount = res1[0]
                result_label.config(text=f"Fine Amount: ₹{fine_amount}", fg="green")
            else:
                result_label.config(text="No fine found for the given Member ID.", fg="red")
    
        # Button to trigger fine check
        check_button = tk.Button(J, text="Check Fine", font=("Arial", 12), command=check_fine)
        check_button.grid(row=2, column=0, columnspan=2, pady=10)
    
        # Back button to close the fine window
        back_button_top_level = tk.Button(J, text="Back", font=("Arial", 12), command=J.destroy)
        back_button_top_level.grid(row=3, column=0, columnspan=2, pady=10)
    
    # Button to check fines in the main window
    check_fine_button = tk.Button(fine_window, text="MEMBER FINE", font=("Arial", 12), command=fine_check)
    check_fine_button.grid(row=k + 1, column=2, columnspan=2, pady=20)
    
    # Back button for the main fine window
    back_button_main = tk.Button(fine_window, text="EXIT", font=("Arial", 12), command=fine_window.destroy)
    back_button_main.grid(row=k + 1, column=0, columnspan=2, pady=20)
    
    # Run the application
    fine_window.mainloop()
```
